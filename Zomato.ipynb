```python
df_country=pd.read_excel('/content/drive/MyDrive/Country-Code[1].xlsx')
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.columns
```

##Transactions

```python
country_names=final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="'%1.2f%%'")
```

Observation:Zomato maximum transcation are from India ,USA and UK

##Ratings

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

Observation:
  1.Not rated count is very high
  2.Maximum number of rating is between 2.5 to 3.4

```python
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().head(5)
```

Observation:
  Maximum numbers of 0 ratings are from indian customers

##Currency

```python
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index().head(5)
```

Observation:
  We can see that the major currency used as Indian rupees

##Delivery

```python
final_df[final_df['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```
